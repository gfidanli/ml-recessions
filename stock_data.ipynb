```python
# Update values in 'quarter' column to have the format: YYYYQ1-4. Example: 1970Q1

def map_quarter(s):
    quarter = ''
    month = s.split('-')[1]
    year = s.split('-')[0]
    
    if month == '01':
        quarter = '1'
    elif month == '04':
        quarter = '2'
    elif month == '07':
        quarter = '3'
    elif month == '10':
        quarter = '4'
        
    x = year + 'Q' + quarter
    
    return x
```

```python
# Convert date to string
from datetime import datetime

data['Date'] = data.Date.apply(lambda x: x.strftime("%Y-%m-%d"))

data['quarter'] = data['Date'].apply(map_quarter)
data.head()
```
